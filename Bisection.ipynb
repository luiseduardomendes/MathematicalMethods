```python
import sympy as sp
x = sp.Symbol('x')
fx = 100*sp.cos(x) - 50*x
```

```python
def bissection(f, x, a, b, tol, plot_iteration=False):
    a_i = a
    b_i = b
    if plot_iteration:
        plot(f, x, a, b, a_i, b_i)
    while (b-a) > tol and f.subs(x, a) * f.subs(x, b) < 0:
        t = (a + b) / 2
        if (f.subs(x, a) * f.subs(x, t)) < 0:
            b = t
        else:
            a = t
        if plot_iteration:
            plot(f, x, a, b, a_i, b_i)
    return (a + b) / 2
        
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot(f, x, a, b, a_i, b_i):
    x_axis = np.linspace(a_i, b_i, 100)
    y_axis = np.array([f.subs(x, i) for i in x_axis])
    plt.grid()
    plt.plot(x_axis, y_axis)
    plot_it(f, x, a, b)
    plt.show()
    
    
def plot_it(f, x, a, b):
    plt.scatter([a, b], [f.subs(x, a), f.subs(x, b)])
```

```python
bissection(fx, x, 0, np.pi/2, 10**-3, plot_iteration=True)
```

```python
n = bissection(sp.cos(x)-x, x, 0, 5, 10^-3)

f = sp.cos(x)

x_axis = np.linspace(0, np.pi/2, 100)
y_axis = np.array([f.subs(x, i) for i in x_axis])
plt.plot(x_axis, y_axis)

f = x

x_axis = np.linspace(0, np.pi/2, 100)
y_axis = np.array([f.subs(x, i) for i in x_axis])
plt.plot(x_axis, y_axis)

f = sp.cos(x) - x

x_axis = np.linspace(0, np.pi/2, 100)
y_axis = np.array([f.subs(x, i) for i in x_axis])
plt.plot(x_axis, y_axis)


x_axis = np.array([n, n])
y_axis = np.array([-1.5, 1.5])
plt.plot(x_axis, y_axis)

plt.grid()
plt.show()
```
