```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.Symbol('x')
fx = sp.cos(x)
```

```python
def fixed_dot_it(f, x, x0, tol):
    plot(f, x, x0, 0)
    xn = float(f.subs(x, x0)) + x0
    plot(f, x, xn, xn - x0)
    while xn - x0 > tol and xn != float('inf'):
        x0 = xn
        xn = float(f.subs(x, xn)) + xn
        
        plot(f, x, xn, xn - x0)
    return xn
```

```python
def plot(f, x, x0, erro):
    x_axis = np.linspace(x0 - 1, x0 + 1, 100)
    y_axis = np.array([f.subs(x, i) for i in x_axis])
    plt.grid()
    plt.plot(x_axis, y_axis)
    plt.title(f'x0 = {x0}, |x - x0| = {erro}')
    plt.scatter(x0, f.subs(x, x0))
    plt.show()
    
```

```python
fixed_dot_it(fx, x, 1, 0.005)
```
