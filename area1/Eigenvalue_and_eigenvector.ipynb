```python
import sympy as sp
import matplotlib.pyplot as plt
A = sp.Matrix([[1, 2],[0,3]]).T
A
```

```python
x1 = sp.Matrix([[-1],[-1]])
x1
```

```python
def plot(a, b):
    plt.grid()
    plt.scatter([0, a],[0, b])
    plt.plot([0, a],[0, b])
    plt.show()
```

```python

def eigenvalues_calculation(A, x1, TOL):
    origin = [[0, 0], [0, 0]]
    x = x1
    
    plot(x[0, 0] , x[1, 0])
    
    y_prev = float((x.T * A * x).norm())
    x = A * x / (A * x).norm()
    y = float((x.T * A * x).norm())
    
    plot(x[0, 0] , x[1, 0])
    
    while(abs(y_prev - y) > TOL):
        y_prev = y
        x = (A * x / (A * x).norm()).evalf()
        y = float((x.T * A * x).norm())
        
        plot(x[0, 0] , x[1, 0])
    return y_prev, x
```

```python
y, x = eigenvalues_calculation(A, x1, 10**(-3))
y
```
