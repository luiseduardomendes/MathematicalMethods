```python
def plot(f, min_x, max_x, x0 = None):
    x_axis = np.linspace(min_x, max_x, 1000)
    y_axis = []
    for x_ in x_axis:
        y_ = f.subs(x, x_)
        y_axis.append(y_)
    
    plt.grid()
    plt.plot(x_axis, y_axis)
    
    if x0 != None:
        f_1 = f.diff(x).subs(x, x0) * (x - x0) + f.subs(x, x0)
        y_axis = []
        for x_ in x_axis:
            y_ = f_1.subs(x, x_)
            y_axis.append(y_)
        plt.plot(x_axis, y_axis)
        plt.scatter([x0], [float(f.subs(x, x0))])
    
    
    
    plt.show()
```

```python
import math
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt

x = sp.symbols('x')

def newton_raphson_method(f, x0, tol, N):
    
    x1 = x0
    x0 = calc_next_x(f, x0)
    #plot(f, -abs(x0 - x1)*5, abs(x0 - x1)*5, x0)
    num_it = 1
    
    while float(abs(x1 - x0)) > tol and num_it < N:
        x1 = x0
        x0 = calc_next_x(f, x0)
        num_it += 1
        #plot(f, -abs(x0 - x1)*5, abs(x0 - x1)*5, x0)
    return x0


def calc_next_x(f, x0):
    fx0 = float(f.subs(x, x0))
    f_x0 = float(f.diff(x).subs(x, x0))
    return x0 - fx0 / f_x0
```
